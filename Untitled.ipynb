```python
import yfinance as yf
```

```python
symbols = ['ACWI', 'EFA', 'EEM', '^GSPC', '^STOXX50E', '^N225', '^HSI']
```

```python
ytd = yf.download(symbols, start='2019-12-28', end='2020-05-31', thread=True)['Adj Close']
```

```python
ytd_ = ytd.fillna(method='ffill').iloc[2:]
```

```python
ytd_.to_excel('market 20200531.xlsx')
```
